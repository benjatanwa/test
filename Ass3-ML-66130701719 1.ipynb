## 2D 100 points 

```python
import pandas as pd

s = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
print('URL:', s)
df = pd.read_csv(s,header=None)
df.tail()

# df = pd.read_csv('iris.data', header=None)
# df.tail()
```

```python
iris_model = PerceptronMe(n_iter=15 , eta=0.01)
iris_model.fit(X, y)
```

## Perceptron- ScikitLearn

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv('iris.data', header=None)
df.tail()
```

```python
y = df.iloc[:, 4].values
X = df.iloc[:, 0:4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state = 123)
```

1. สร้างโมเดล Perceptron โดยใช้ sklearn 
2. ทดสอบโมเดล
3. หาค่าคะแนนความถูกต้อง

```python
model_per = Perceptron(tol=1e-3, random_state=0)
model_per.fit(X_train, y_train)
model_per.score(X_test, y_test)
```

4. บันทึก model ที่สร้างขึ้น 

```python
import pickle
pickle.dump(model_per, open('per_model-66130701719.sav', 'wb'))
```

## สร้างโปรแกรม ass3-app.py


```python
import streamlit as st
import pickle
import numpy as np

model = pickle.load(open('per_model-66130701719.sav', 'rb'))
st.title('Iris Flower Prediction App')

x1 = st.slider('Sepal Length', 0.0, 10.0, 0.1)
x2 = st.slider('Sepal Width', 0.0, 10.0, 0.1)
x3 = st.slider('Petal Length', 0.0, 10.0, 0.1)
x4 = st.slider('Petal Width', 0.0, 10.0, 0.1)

input_data = np.array([[x1, x2, x3, x4]]).reshape(1, -1)

if st.button('Predict'):
    result = model.predict(input_data)
    st.write('The result is:', result)
    st.write('The flower is:', result[0])
```
